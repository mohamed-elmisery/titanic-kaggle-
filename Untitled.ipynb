
## 2Ô∏è‚É£ Data Cleaning

```python
# Count missing values
train.isnull().sum()
```

```python
# we will fill nulls with median age 
train.fillna({"Age": train["Age"].median()}, inplace=True)
```

```python
# Not many missing im Embarked so we will replace it with mode 
train["Embarked"].fillna(train["Embarked"].mode()[0], inplace=True)
```

```python
# -------------------------------
# Handling the 'Cabin' column
# -------------------------------

# 1) Create a new feature: Cabin_flag ‚Üí Did the passenger have a cabin or not?
# .notnull() returns True if a value exists, False if it's missing
# .astype(int) converts True=1 and False=0
train["Cabin_flag"] = train["Cabin"].notnull().astype(int)
```

```python
# 2) Drop the original 'Cabin' column
# Reason: More than 70% of the values are missing, 
# so the column is not reliable to use directly
train.drop("Cabin", axis=1, inplace=True)
```
